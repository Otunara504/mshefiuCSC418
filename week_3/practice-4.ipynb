```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sst.jpg', 0)

#Plotting the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

cropped_img = img[450:700, 200:500]

#Plotting the cropped image
plt.subplot(1, 2, 2)
plt.title("Cropped Image")
plt.imshow(cropped_img)

cv.imwrite('img/cropped_out.jpg', cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# Load image
img = cv.imread('img/sst.jpeg', 0)

# Check if image loaded
if img is None:
    print("Error: Image not found!")
    exit()

# Get image dimensions
height, width = img.shape
print(f"Image size: {width}x{height}")  # Debugging info

# Ensure cropping is within image bounds
y1, y2 = max(0, height - 100), height  # Crop last 100 pixels
x1, x2 = max(0, width - 100), width    # Crop last 100 pixels

cropped_img = img[y1:y2, x1:x2]

# Check if cropping is valid
if cropped_img.size == 0:
    print("Error: Cropped image is empty! Adjust crop size.")
    exit()

# Plot original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')

# Plot cropped image
plt.subplot(1, 2, 2)
plt.title("Cropped Image")
plt.imshow(cropped_img, cmap='gray')

plt.show()

# Save cropped image
cv.imwrite('img/cropped_out.jpg', cropped_img)
print("Cropped image saved successfully!")
```
