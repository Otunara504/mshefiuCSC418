```python
# Import the libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('img/sst-foyer.jpeg')

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

# Inverse by subtracting from 255
inverse_image = 255 - image

# Save the image
cv2.imwrite('img/inverse_image.jpeg', inverse_image)

# Plot the blurred image
plt.subplot(1, 2, 2)
plt.title("Inverse colour")
plt.imshow(inverse_image)
plt.show()
```
