```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def filterPicker():
    username = input("Enter username: ")
    password = input("Enter matno: ")

    credentials = [username, password]

    # Dict for image paths
    image_paths = {
        "AD": 'img/students/adaobi.jpg',
        "BI": 'img/students/bimpe.jpg',
        "DA": 'img/students/daniel.jpg',
        "DU": 'img/students/dumebi.jpg',
        "LO": 'img/students/love.jpg',
        "OY": 'img/students/oyinda.jpg',
        "PR": 'img/students/precious.jpg',
        "TA": 'img/students/tabitha.jpg',
        "TE": 'img/students/testimony.jpg',
        "TI": 'img/students/tireni.jpg'
    }

    # Dict for username and matno
    records = {
        "AD": ["adaobi", "0001"], 
        "BI": ["bimpe", "0010"], 
        "DA": ["daniel", "0011"], 
        "DU": ["dumebi", "0101"], 
        "LO": ["love", "0110"], 
        "OY": ["oyinda", "0111"], 
        "PR": ["precious", "1110"], 
        "TA": ["tabitha", "1101"], 
        "TE": ["testimony", "1011"],
        "TI": ["tireni", "1111"]
    }

    for i in records:
        if records[i] == credentials:
            print(f"Login Successful, Welcome {username}")

            
            image_path = image_paths[i]
            image = cv2.imread(image_path)

            plt.subplot(1, 2, 1)
            plt.title("Original")
            plt.imshow(image)

            # Filter Picker
            filter = input("""Pick a letter based on filter choice: 
            a. Split colours
            b. Brighten image
            c. Sharpen image
            d. Blur image
            e. Scale image
            f. Invert image
            Enter your choice: """)

            if filter == "a":
                r, g, b = cv2.split(image)
        
                # Plotting the split colours
                plt. subplot(1, 2, 2)
                plt.title("Red")
                plt.imshow(r)
                plt.show()

                plt. subplot(1, 2, 3)
                plt.title("Green")
                plt.imshow(g)
                plt.show()

                plt. subplot(1, 2, 4)
                plt.title("Blue")
                plt.imshow(b)
                plt.show()

            elif filter == "b":        
                # Adjust the brightness and constrast
                brightness = 5
                constrast = 1.5
                image2 = cv2.addWeighted(image, constrast, np.zeros(image.shape, image.dtype), 0, brightness)

                # Plot the constrast image
                plt.subplot(1, 2, 2)
                plt.title("Brightned Image")
                plt.imshow(image2)
                plt.show()

            elif filter == "c":
                kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

                # Sharpen the image
                sharpened_image = cv2.filter2D(image, -1, kernel)

                # Plot the sharpened image
                plt. subplot(1, 2, 2)
                plt.title("Sharpened Image")
                plt.imshow(sharpened_image)
                plt.show()

            elif filter == "d":
                # Remove noise using a median filter
                blurred_image = cv2.medianBlur(image, 15)

                # Plot the blurred image
                plt.subplot(1, 2, 2)
                plt.title("Blurred Image")
                plt.imshow(blurred_image)
                plt.show()

            elif filter == "e":         
                # Scale the image by a factor of 2 along both axes
                scaled_image = cv2.resize(image, None, fx=2, fy=2)

                # Plot the blurred image
                plt.subplot(1, 2, 2)
                plt.title("Scaled Image")
                plt.imshow(scaled_image)
                plt.show()

            elif filter == "f":       
                # Inverse by subtracting from 255
                inverse_image = 255 - image

                # Plot the blurred image
                plt.subplot(1, 2, 2)
                plt.title("Inversed Image")
                plt.imshow(inverse_image)
                plt.show()
            break
            
    else:
        print("Login Failed")

    

filterPicker()
```
