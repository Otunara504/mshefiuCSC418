```python
# Import the necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('img/sst.jpeg')

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

# Adjust the brightness and constrast
brightness = 5
constrast = 1.5
image2 = cv2.addWeighted(image, constrast, np.zeros(image.shape, image.dtype), 0, brightness)

# Save the image
cv2.imwrite('img/constrast_image.jpeg', image2)
# Plot the constrast image
plt.subplot(1, 2, 2)
plt.title("Brightness & constrast")
plt.imshow(image2)
plt.show()
```
